```python
import os
import glob
import PyPDF2
import chromadb
from chromadb.config import Settings
from langchain_chroma import Chroma
from langchain_community.document_loaders import PyPDFLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.llms import OpenAI
from langchain.chains import RetrievalQA
from langchain.prompts import PromptTemplate
from langchain.schema import Document
import getpass

os.environ["OPENAI_API_KEY"] = getpass.getpass("Enter your OpenAI API key: ")
```

```python
data_dir = "data/"
print("Contents of data directory:", os.listdir(data_dir))


if os.path.exists("data/files/"):
    print("Contents of data/files:", os.listdir("data/files/"))
else:
    print("data/files directory doesn't exist")


pdf_files = []
for root, dirs, files in os.walk(data_dir):
    for file in files:
        if file.endswith(".pdf"):
            pdf_files.append(os.path.join(root, file))

print(f"Found {len(pdf_files)} PDF files:")
for pdf in pdf_files:
    print(f"  - {pdf}")


if len(pdf_files) == 0:
    print("No PDFs found. Creating a sample text file for testing...")
    sample_text = """
    InfoTech College offers various programs in technology and computer science.
    We have courses in programming, web development, data science, and cybersecurity.
    Our college was established in 2005 and has since graduated over 5000 students.
    We offer both online and on-campus programs with flexible scheduling options.
    Contact us at admissions@infotechcollege.com for more information.
    
    Programs Offered:
    - Computer Science Degree (4 years)
    - Web Development Diploma (2 years)
    - Data Science Certificate (1 year)
    - Cybersecurity Bootcamp (6 months)
    
    Admission Requirements:
    - High school diploma or equivalent
    - Completed application form
    - Application fee
    - Personal statement
    
    Tuition and Fees:
    - Full-time tuition: $5,000 per semester
    - Part-time tuition: $400 per credit hour
    - Technology fee: $150 per semester
    - Application fee: $50 (non-refundable)
    
    Financial Aid:
      We offer various financial aid options including scholarships, grants, and student loans.
    Contact our financial aid office at financialaid@infotechcollege.com for more information.
    
    Campus Location:
    123 Tech Avenue, Innovation City, IC 12345
    Phone: (555) 123-4567
    Email: info@infotechcollege.com
    """
    

    os.makedirs("data", exist_ok=True)
    with open("data/sample_info.txt", "w") as f:
        f.write(sample_text)
    pdf_files = ["data/sample_info.txt"]


def extract_text_from_file(file_path):
    text = ""
    try:
        if file_path.endswith(".pdf"):
            with open(file_path, 'rb') as file:
                reader = PyPDF2.PdfReader(file)
                for page in reader.pages:
                    text += page.extract_text() + "\n"
        elif file_path.endswith(".txt"):
            with open(file_path, 'r', encoding='utf-8') as file:
                text = file.read()
        else:
            print(f"Unsupported file type: {file_path}")
    except Exception as e:
        print(f"Error processing {file_path}: {str(e)}")
    return text


documents = []
for file_path in pdf_files:
    try:
        text = extract_text_from_file(file_path)
        if text.strip(): 
            documents.append(Document(page_content=text, metadata={"source": file_path}))
            print(f"Processed: {file_path}")
        else:
            print(f"No text extracted from: {file_path}")
    except Exception as e:
        print(f"Error processing {file_path}: {str(e)}")

print(f"Extracted text from {len(documents)} documents")
```

```python

text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=1000,
    chunk_overlap=200,
    length_function=len
)

chunks = text_splitter.split_documents(documents)
print(f"Split into {len(chunks)} chunks")


if chunks:
    print("\nSample chunk:")
    print(chunks[0].page_content[:500] + "...")
```

```python

from langchain_openai import OpenAIEmbeddings
embeddings = OpenAIEmbeddings()


chroma_client = chromadb.PersistentClient(path="./chroma_db")


try:
    collection = chroma_client.get_collection(name="infotech_college")
    print("Using existing collection")
except:
    collection = chroma_client.create_collection(name="infotech_college")
    print("Created new collection")


doc_ids = [f"doc_{i}" for i in range(len(chunks))]
doc_texts = [chunk.page_content for chunk in chunks]
doc_metadatas = [chunk.metadata for chunk in chunks]

collection.add(
    ids=doc_ids,
    documents=doc_texts,
    metadatas=doc_metadatas
)

print("Vector database created and persisted")
```

```python
llm = OpenAI(temperature=0)


vectorstore = Chroma(
    client=chroma_client,
    collection_name="infotech_college",
    embedding_function=embeddings
)

retriever = vectorstore.as_retriever(search_kwargs={"k": 3})


prompt_template = """You are a helpful assistant for InfoTech College. Use the following pieces of context to answer the question at the end. 
If you don't know the answer, just say that you don't know, don't try to make up an answer.

{context}

Question: {question}
Answer:"""

PROMPT = PromptTemplate(
    template=prompt_template, input_variables=["context", "question"]
)


qa_chain = RetrievalQA.from_chain_type(
    llm=llm,
    chain_type="stuff",
    retriever=retriever,
    chain_type_kwargs={"prompt": PROMPT},
    return_source_documents=True
)
```

```python
def local_embedding_function(texts):
    from sentence_transformers import SentenceTransformer
    model = SentenceTransformer('all-MiniLM-L6-v2')
    return model.encode(texts).tolist()


def chatbot(query):
    try:
        
        results = collection.query(
            query_texts=[query],
            n_results=3
        )
        
        
        context = "\n".join(results['documents'][0])
        
        
        response = f"I found this information that might help:\n\n{context}\n\nFor more specific details, please contact InfoTech College directly."
        
        
        sources = list(set([meta['source'] for meta in results['metadatas'][0]]))
        
        return f"{response}\n\nSources: {sources}"
    except Exception as e:
        return f"I encountered an error while processing your query. Please try again or contact the college directly. Error: {str(e)}"


test_questions = [
    "What programs does InfoTech College offer?",
    "How can I contact the college?",
    "What is the tuition cost?",
]

for question in test_questions:
    print(f"Q: {question}")
    response = chatbot(question)
    print(f"A: {response}\n{'-'*50}")


def run_chatbot():
    print("Welcome to InfoTech College Chatbot! Type 'exit' to end the conversation.")
    print("You can ask me about programs, admissions, or any other information about InfoTech College.")
    
    while True:
        user_input = input("\nYou: ")
        
        if user_input.lower() in ['exit', 'quit', 'bye']:
            print("Chatbot: Thank you for chatting with me. Have a great day!")
            break
            
        if user_input.strip() == "":
            continue
            
        response = chatbot(user_input)
        print(f"\nChatbot: {response}")


run_chatbot()
```
